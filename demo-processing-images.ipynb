### Imports

```python
import imageio
import matplotlib.pyplot as plt
import augmentation as aug
```

### Load the image

```python
filename = 'Dataset/Train/COVID/COVID-1.png'
# filename = 'Dataset/Test/NON_COVID/Lung_Opacity-8.png'
img = imageio.imread(filename)
```

### Noise insertion

```python
r_img = aug.add_noise(img, 5, 5)
plt.imshow(r_img, cmap="gray")
```

### Contrast adjustment

```python
r_img = aug.adjust_contrast(img, 1.1)
plt.imshow(r_img, cmap="gray")
```

### Sharpness adjustment

```python
r_img = aug.adjust_sharpness(img, 0.1, 1.5, 7.5)
plt.imshow(r_img, cmap="gray")
```

### Rotation

```python
r_img = aug.rotate(img, 15)
plt.imshow(r_img, cmap="gray")
```
